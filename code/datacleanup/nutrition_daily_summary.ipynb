```python
import pandas as pd, re
import io
import csv

df = pd.read_csv("/Users/clee/Documents/Lab/mlife/data/raw/nutrition_daily_summary.csv")
```

```python
column_zero_counts = df.eq(0).sum()
column_zero_pct = df.eq(0).mean().mul(100)

column_zero_summary = (
    pd.DataFrame({
        "zero_count": column_zero_counts,
        "zero_pct_of_rows": column_zero_pct,
    })
    .sort_values("zero_pct_of_rows", ascending=False)
)

display(column_zero_summary.style.format({"zero_pct_of_rows": "{:.2f}%"}))

```

```python
non_zero_group_counts = (
    df.melt(id_vars="Group", var_name="column", value_name="value")
      .query("value != 0")
      .groupby(["column", "Group"])
      .size()
      .unstack(fill_value=0)
      .sort_index()
)

display(non_zero_group_counts)
```
